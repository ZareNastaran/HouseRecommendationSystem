```python
from IPython.display import display
missing_values = data.isnull().sum()
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(missing_values)
```

```python
features_to_delete = ['Alley', 'FireplaceQu', 'PoolQC','Fence','MiscFeature']
data = data.drop(features_to_delete, axis=1)
data = data.drop("Id", axis=1)
data
```

```python
missing_values_per_record = data.isnull().sum(axis=1)
rows_with_missing_values = data[data.isnull().sum(axis=1) >= 10]
print(rows_with_missing_values)
```

```python
indices_to_delete = rows_with_missing_values.index
dataoriginal = data.drop(indices_to_delete)
data = data.drop(indices_to_delete)
data
```

```python
categorical_features = ['SaleType','MSZoning', 'Street','LotShape','LandContour','Utilities','LotConfig','LandSlope','Neighborhood','Condition1','Condition2','BldgType','HouseStyle','RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType','ExterQual','ExterCond','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Heating','HeatingQC','CentralAir','Electrical','KitchenQual','Functional','GarageType','GarageFinish','GarageQual','GarageCond','PavedDrive','SaleCondition']
unique_values = dataoriginal[categorical_features].nunique()
unique_values
```

```python
for feature in categorical_features:
    unique_values = data[feature].unique()
    print("Unique values for", feature, ":")
    print(unique_values)
    print()
```

```python
ranking_dict = {
    'nan':3,
    'Po': 1,
    'Fa': 2,
    'Ta': 3,
    'Gd': 4,
    'Ex': 5
}

features_to_convert = ['BsmtQual', 'BsmtCond', 'HeatingQC','KitchenQual','GarageQual','GarageCond']

for feature in features_to_convert:
    data[feature] = data[feature].map(ranking_dict)
    
data
```

```python
categorical_features = [x for x in categorical_features if x not in features_to_convert]

for column in data.columns:
    if column not in categorical_features:
        if data[column].dtype != 'object':
            average = np.nanmean(data[column])
            data[column].fillna(average, inplace=True)
            
missing_values = data.isnull().sum()
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(missing_values)
```

```python
for feature in categorical_features:
    most_frequent_value = data[feature].mode().iloc[0]
    data[feature].fillna(most_frequent_value, inplace=True)
```

```python
missing_values = data.isnull().sum()
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(missing_values)
```

```python
duplicates = data.duplicated()
print(data[duplicates])
```

```python
data = pd.get_dummies(data, columns=categorical_features)
```

```python
feature_names = list(data.columns)
num_features = list(set(feature_names) - set(categorical_features))
num_features
```

```python
#for the first try we can test MinMaxScaler but there are Z-score and RobustScaling that we can test in the improvement phase
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[num_features] = scaler.fit_transform(data[num_features])
```

```python
data
```

```python

from sklearn.metrics.pairwise import cosine_similarity

cosine_sim_matrix = cosine_similarity(data)

cosine_sim_data = pd.DataFrame(cosine_sim_matrix, index=data.index, columns=data.index)
```

```python
record_number = int(input("Enter the record number (row index): "))

input_row_scores = cosine_sim_data.iloc[record_number]

top_similar_records = input_row_scores.sort_values(ascending=False)[1:6]

similar_records = data.iloc[top_similar_records.index]

print(similar_records)
```

```python
selected_records = [record_number] + top_similar_records.index.tolist()

selected_data = dataoriginal.iloc[selected_records]

print(selected_data)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = text.lower()

    tokens = nltk.word_tokenize(text)

    filtered_tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]

    preprocessed_text = " ".join(filtered_tokens)

    return preprocessed_text

des['Description'] = des['Description'].apply(preprocess_text)
```
