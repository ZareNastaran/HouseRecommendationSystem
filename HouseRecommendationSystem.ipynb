```python
categorical_features = ['MSZoning', 'Street','LotShape','LandContour','Utilities','LotConfig','LandSlope','Neighborhood','Condition1','Condition2','BldgType','HouseStyle','RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType','ExterQual','ExterCond','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Heating','HeatingQC','CentralAir','Electrical','KitchenQual','Functional','GarageType','GarageFinish','GarageQual','GarageCond','PavedDrive','SaleCondition']
unique_values = data[categorical_features].nunique()
unique_values
```

```python
for feature in categorical_features:
    unique_values = data[feature].unique()
    print("Unique values for", feature, ":")
    print(unique_values)
    print()
```

```python
ranking_dict = {
    'nan':3,
    'Po': 1,
    'Fa': 2,
    'Ta': 3,
    'Gd': 4,
    'Ex': 5
}

features_to_convert = ['BsmtQual', 'BsmtCond', 'HeatingQC','KitchenQual','GarageQual','GarageCond']

for feature in features_to_convert:
    data[feature] = data[feature].map(ranking_dict)
    
data
```

```python
categorical_features = [x for x in categorical_features if x not in features_to_convert]

for column in data.columns:
    if column not in categorical_features:
        if data[column].dtype != 'object':
            average = np.nanmean(data[column])
            data[column].fillna(average, inplace=True)
            
missing_values = data.isnull().sum()
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(missing_values)
```

```python
for feature in categorical_features:
    most_frequent_value = data[feature].mode().iloc[0]
    data[feature].fillna(most_frequent_value, inplace=True)
```

```python
missing_values = data.isnull().sum()
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(missing_values)
```
